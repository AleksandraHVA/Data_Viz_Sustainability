## Calculating the overall sustainability score

```python
selected_activities = [
    'Activities to save energy and/or water',
    'Activities to reduce waste',
    'Activities to travel less or more sustainable',
    'Activities to buy less or second-hand',
    'Activities to rent instead of buying goods'
]

df1 = df_open_q[df_open_q['Variable group'].isin(selected_activities)]
```

```python
# choosing only "Selected". otherwise the score will be the same across all repondents
df1 = df1[df1['Value label'] != 'Not selected']
```

Now we define several functions according to the activities.

```python
def calculate_energy_water_score(activity):
    if activity == 'None of these':
        return 0
    elif activity == 'I buy as little as possible, unless there is no other way':
        return 2
    elif activity == 'I first look on the internet to see if I can buy something second-hand (e.g. on Marktplaats.nl or other website)':
        return 3
    elif activity == 'I first look at the thrift shop to see if I can buy something second-hand':
        return 3
    elif activity == 'I first ask family or friends if I can borrow/acquire the product I need, before I buy it new':
        return 2
    elif activity == 'I buy refurbished products (e.g. phones, computers) whenever possible':
        return 3
    elif activity == 'I make products available through sharing platforms (e.g. Peerby, Olio, Nextdoor, groups on Facebook) for others to buy or borrow':
        return 1
    elif activity == 'I use sharing platforms (e.g. Peerby, Olio, Nextdoor, groups on Facebook) to buy or to borrow products':
        return 2
    else:
        return 0

def calculate_reduce_waste_score(activity):
    if activity == 'None of these':
        return 0
    elif activity == 'Bring my own bag when I go shopping':
        return 3
    elif activity == 'Do not buy pre-packed products':
        return 2
    elif activity == 'Growing your own fruit and vegetables':
        return 2
    elif activity == 'Consciously buying products close to their sell-by date':
        return 2
    elif activity == 'Buying as little as possible, only the bare essentials':
        return 3
    elif activity == 'Buy second-hand products as much as possible':
        return 3
    elif activity == 'Avoid releasing microplastics into the environment (e.g. via filter for washing machine, microplastic-free)':
        return 3
    elif activity == 'Throw away as little food as possible and/or save leftovers for another meal':
        return 3
    elif activity == 'Have defective products (clothes, shoes, electronics, etc.) repaired if possible, rather than disposing of them':
        return 3
    elif activity == 'Bring your own containers/bags for fresh produce such as bread and vegetables when I go shopping':
        return 2
    elif activity == 'Buy products that generate as little waste as possible':
        return 2
    elif activity == 'Make your own products (e.g. toothpaste, shampoo, detergent)':
        return 2
    elif activity == 'Buy reusable products instead of disposable products':
        return 3
    elif activity == 'Composting vegetable and garden waste':
        return 3
    elif activity == 'Separate waste to encourage recycling (e.g. plastic, paper, clothing, glass)':
        return 2
    elif activity == 'Giving things a second life by selling, giving away or swapping them':
        return 3
    else:
        return 0

def calculate_mobility_score(activity):
    if activity == 'None of these':
        return 0
    elif activity == 'I consciously do not own a car, scooter, or motorbike':
        return 3
    elif activity == 'I consciously fly less':
        return 2
    elif activity == 'I rent/lend a car, scooter, or motorbike when I need one':
        return 2
    elif activity == 'I consciously drive a more fuel-efficient car (e.g., smaller car, hybrid car, electric car, gas car)':
        return 2
    elif activity == 'When I use the car, I consciously think about combining trips (e.g., shopping and commuting)':
        return 2
    elif activity == 'For travel within Europe, I prefer to travel by train and/or bus instead of by car or plane':
        return 3
    elif activity == 'I drive an electric car (as much as possible)':
        return 3
    elif activity == 'I travel by public transport as much as possible':
        return 2
    elif activity == 'I walk and cycle as much as possible':
        return 3
    elif activity == 'I consciously don''t fly (anymore)':
        return 3
    elif activity == 'I choose CO2 compensation when I fly':
        return 2
    else:
        return 0

def calculate_buy_less_second_hand_score(activity):
    if activity == 'None of these':
        return 0
    elif activity == 'I buy as little as possible, unless there is no other way':
        return 2
    elif activity == 'I buy refurbished products (e.g. phones, computers) whenever possible':
        return 3
    elif activity == 'I first look on the internet to see if I can buy something second-hand (e.g. on Marktplaats.nl or other website)':
        return 2
    elif activity == 'I first look at the thrift shop to see if I can buy something second-hand':
        return 3
    elif activity == 'I first ask family or friends if I can borrow/acquire the product I need, before I buy it new':
        return 2
    elif activity == 'I use sharing platforms (e.g. Peerby, Olio, Nextdoor, groups on Facebook) to buy or to borrow products':
        return 1
    elif activity == 'I make products available through sharing platforms (e.g. Peerby, Olio, Nextdoor, groups on Facebook) for others to buy or borrow':
        return 1
    else:
        return 0

def calculate_save_energy_score(activity):
    if activity == 'None of these':
        return 0
    elif activity == 'Consciously turn on the heating lower or shorter':
        return 2
    elif activity == 'Via a green roof (e.g. sedum roof)':
        return 3
    elif activity == 'I live in an energy-efficient home':
        return 3
    elif activity == 'Using low-energy light bulbs or LED lighting':
        return 3
    elif activity == 'I turn off lights in rooms where nobody is present':
        return 2
    elif activity == 'I dry my laundry on a clothes rack/washing line instead of using a tumble dryer':
        return 3
    elif activity == 'I consciously wash at low temperatures':
        return 3
    elif activity == 'I take energy from solar panels elsewhere':
        return 2
    elif activity == 'Using smart lighting (which turns off when no one is in the room)':
        return 3
    elif activity == 'Showering or bathing less often':
        return 2
    elif activity == '(Rain) water collection for (re)use':
        return 3
    elif activity == 'I buy electrical appliances with the lowest possible energy consumption':
        return 3
    elif activity == 'Using energy and/or water-saving products (e.g. shower head or programs on dishwasher)':
        return 3
    elif activity == 'I have had my house insulated/I live in a well-insulated house':
        return 3
    elif activity == 'None of these':
        return 0
    elif activity == 'Via a wood or pellet stove (or fireplace)':
        return 2
    elif activity == 'Consciously taking short(er) and/or cold showers':
        return 2
    elif activity == 'I use green energy':
        return 3
    elif activity == 'Using a smart energy meter to monitor my energy consumption':
        return 3
    elif activity == 'Via solar panels on my house':
        return 3
    elif activity == 'Via a heat pump':
        return 3
    else:
        return 0
```

```python
# Creating new columns for each activity category score
df1['Energy_Water_Score'] = df1['Variable options/statements'].apply(calculate_energy_water_score)
df1['Reduce_Waste_Score'] = df1['Variable options/statements'].apply(calculate_reduce_waste_score)
df1['Mobility_Score'] = df1['Variable options/statements'].apply(calculate_mobility_score)
df1['Buy_Less_Second_Hand_Score'] = df1['Variable options/statements'].apply(calculate_buy_less_second_hand_score)
df1['Save_Energy_Score'] = df1['Variable options/statements'].apply(calculate_save_energy_score)

# Displaying the resulting DataFrame
(df1[['ResponseId', 'Energy_Water_Score', 'Reduce_Waste_Score', 'Mobility_Score', 'Buy_Less_Second_Hand_Score', 'Save_Energy_Score']])
```

Now you can see that we have the score per question so we have a repeating ID, so we need to aggregate it. But first we are calculating the overall sustainability score by summing all of the mini scores

```python
# Defining a list of category score columns
category_score_columns = ['Energy_Water_Score', 'Reduce_Waste_Score', 'Mobility_Score', 'Buy_Less_Second_Hand_Score', 'Save_Energy_Score']

# Creating a new column 'Overall_Sustainability_Score' by summing up scores across categories for each respondent
df1['Overall_Sustainability_Score'] = df1[category_score_columns].sum(axis=1)

# Display the resulting DataFrame
(df1[['ResponseId', 'Overall_Sustainability_Score']])
```

And now we aggregate per ID

```python
overall_score_per_id = df1.groupby('ResponseId').agg({'Energy_Water_Score': 'sum',
                                                      'Reduce_Waste_Score': 'sum',
                                                      'Mobility_Score': 'sum',
                                                      'Buy_Less_Second_Hand_Score': 'sum',
                                                      'Save_Energy_Score': 'sum'}).sum(axis=1).reset_index(name='Overall_Sustainability_Score')

overall_score_per_id
```

Afterwards, we wanted to check the upper and lower limits in order to create the threshold - e.g. "Unprepared", "Static", "Prepared"

```python
overall_score_per_id.describe()
```

```python
# Defining the categories based on percentiles
unprepared_threshold = overall_score_per_id['Overall_Sustainability_Score'].quantile(0.25)
prepared_threshold = overall_score_per_id['Overall_Sustainability_Score'].quantile(0.75)

# Creating a new column 'Category' based on the defined thresholds
overall_score_per_id['Category'] = pd.cut(overall_score_per_id['Overall_Sustainability_Score'], bins=[-float('inf'), unprepared_threshold, prepared_threshold, float('inf')],
                        labels=['Unprepared', 'Static', 'Prepared'])

overall_score_per_id
```

# Perceived vs real sustainability score

We thought it would be interesting to see if there is a big of a difference between the sustainability perception people have of themselves and their real sustainability score

We have the following statement "If you were allowed to give a rating for how sustainable your personal lifestyle is, what rating would you give yourself on a scale of 1 (very inadequate, I do nothing at all) to 10 (excellent, I live as sustainably as possible)?" and we could use it to understand what people think of their sustainability level - same logic with the AI Act (perceived vs real value)

```python
selected_statements = [
    'If you were allowed to give a rating for how sustainable your personal lifestyle is, what rating would you give yourself on a scale of 1 (very inadequate, I do nothing at all) to 10 (excellent, I live as sustainably as possible)?'
]

perceived = df_open_q[df_open_q['Variable description'].isin(selected_statements)] # i make a dataframe with only that statement
```

```python
merged = pd.merge(perceived, overall_score_per_id, on='ResponseId', how='inner')
```

```python
# Making a datframe with only the needed calumns 
selected_columns = ['ResponseId', 'Overall_Sustainability_Score', 'Value label']

selected_df_1 = merged[selected_columns]
```

So we see that for some reason someone just wrote "Selected" so I am replacing that we the median value

```python
# Converting 'Value label' to numeric, replacing any non-numeric values with NaN
selected_df_1['Value label'] = pd.to_numeric(selected_df_1['Value label'], errors='coerce')

# Calculating the median of the 'Value label' column (excluding NaN values)
median_value = selected_df_1['Value label'].median()

# Replace "Selected" with the median value in the 'Value label' column
selected_df_1['Value label'] = selected_df_1['Value label'].replace("Selected", median_value)

# Display the modified DataFrame
print(selected_df_1)
```
